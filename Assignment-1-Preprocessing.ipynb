```python
# Clean Categorical Entries 
# Standardize Activity Level values
activity_map = {
    'Highly_Active': 'Highly Active',
    'highly_active': 'Highly Active',
    'Seddentary': 'Sedentary',
    'Actve': 'Active',
    'Active': 'Active',
    'Sedentary': 'Sedentary',
    'highly active': 'Highly Active', 
    'active': 'Active',
    'sedentary': 'Sedentary'}
ds['Activity Level'] = ds['Activity Level'].replace(activity_map)
# Convert all the entries into one standard format. Eg - Highly_Active to Highly Active
```

```python
# Convert to Proper Types and Handle Errors 
numeric_cols = ['Heart Rate (BPM)', 'Blood Oxygen Level (%)', 'Step Count', 'Sleep Duration (hours)']
for col in numeric_cols:
    ds[col] = pd.to_numeric(ds[col], errors='coerce')
```

```python
ds
```

```python
# Drop duplicates
ds.drop_duplicates(inplace=True)
```

```python
ds
```

```python
ds.head(100)
```

```python
# Drop missing values
ds = ds.dropna(subset=['Heart Rate (BPM)', 'Blood Oxygen Level (%)', 'Step Count', 'Sleep Duration (hours)'], how='any')
```

```python
ds
```

```python
from sklearn.preprocessing import MinMaxScaler
# Normalize Heart Rate
ds = ds.copy()
scaler = MinMaxScaler()
ds.loc[:, 'Heart Rate Norm'] = scaler.fit_transform(ds[['Heart Rate (BPM)']])

# Feature engineering
ds.loc[:, 'Steps_per_Hour_Sleep'] = ds.loc[:, 'Step Count'] / ds.loc[:, 'Sleep Duration (hours)']
```

```python
ds
```

```python
# Convert to numeric 
pd.set_option('future.no_silent_downcasting', True)
ds.loc[:, 'Stress Level'] = pd.to_numeric(ds['Stress Level'], errors='coerce')

# Calculate median 
median_val = ds.loc[:, 'Stress Level'].median()

# Impute missing values with median
ds.loc[:, 'Stress Level'] = ds['Stress Level'].fillna(median_val).infer_objects(copy=False)
```

```python
# Discretize Heart Rate into bins (Low, Normal, High)
bins = [0, 60, 100, np.inf]
labels = ['Low', 'Normal', 'High']
ds.loc[:, 'Heart_Rate_Category'] = pd.cut(ds['Heart Rate (BPM)'], bins=bins, labels=labels)

stress_bins = [0, 3, 7, 10]
stress_labels = ['Low', 'Medium', 'High']
ds.loc[:, 'Stress_Level_Category'] = pd.cut(ds['Stress Level'], bins=stress_bins, labels=stress_labels, include_lowest=True)
```

```python
ds
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
ds.loc[:, ['hr_norm', 'ox_norm', 'steps_norm', 'sleep_norm']] = scaler.fit_transform(ds[['Heart Rate (BPM)', 'Blood Oxygen Level (%)', 'Step Count', 'Sleep Duration (hours)']])

ds.loc[:, 'Health_Score'] = (
    0.25 * ds['hr_norm'] +
    0.25 * ds['ox_norm'] +
    0.25 * ds['steps_norm'] +
    0.25 * ds['sleep_norm']
)
print(ds.loc[:, ['Health_Score']].describe())
```

```python
# Replace empty strings with NaN
ds['Activity Level'] = ds['Activity Level'].replace('', np.nan)

# Fill NaN with 'NA'
ds['Activity Level'] = ds['Activity Level'].fillna('NA')
```

```python
ds.isna().sum()
```

```python
ds = ds.dropna(subset=['User ID'])
```

```python
ds.isna().sum()
```

```python
# Perform one-hot encoding on Activity Levels
one_hot_encoded = pd.get_dummies(ds['Activity Level'], prefix='Activity')

ds = ds.join(one_hot_encoded)
```
